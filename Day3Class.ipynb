```python
#Array concatenation. In 2D axis=0(row) and 1(column)
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])

arr2 = np.array([[7, 8, 9]])
z = np.concatenate((arr1, arr2), axis=0)
z
```

```python
arr2 = np.array([[7],
                 [8]]) #Shape(2,1)
z = np.concatenate((arr1, arr2), axis=1)
z
```
